#### Exercise 4.01 – underlying data types and conversion

```python
import pandas as pd
```

```python
file_url = 'files/Chapter04/Data/retail_purchase.csv'
data_frame = pd.read_csv(file_url)
data_frame.head()
```

```python
data_frame.tail()
```

```python
data_frame.info()
```

```python
# after seeing the dtypes for our dataframe, we need to make some changes
# let us first convert the  Date of Purchase column into datetime

data_frame['Date of Purchase'] = pd.to_datetime(data_frame['Date of Purchase'],
                                                format='%d/%m/%y')
data_frame['Date of Purchase']
```

```python
# now we will remove the € symbol using other method, note the data type will continue being object
data_frame['Total Price'] = data_frame['Total Price'].str[1:]
data_frame['Total Price']
```

```python
# converting the Total Price into a float 
data_frame['Total Price'] = data_frame['Total Price'].astype('float')
data_frame['Total Price']
```

```python
# to remove the "lb" from the product weight column, we will use a similar method to the one used to remove the € from the total price

data_frame['Product Weight'] = data_frame['Product Weight'].str[:-2]
data_frame['Product Weight']
```

```python
# now we can convert the column into a float dtype
data_frame['Product Weight'] = data_frame['Product Weight'].astype('float')
data_frame['Product Weight']
```

```python
# use the unique() function to find every unique value from the Product Name column:
data_frame['Product Name'].unique()
```

```python
# converting product name into category dtype
data_frame['Product Name'] = data_frame['Product Name'].astype('category')
data_frame['Product Name']
```

```python
data_frame['Retail shop name'].unique()
```

```python
data_frame['Retail shop name'] = data_frame['Retail shop name'].astype('category')

data_frame['Retail shop name']
```

```python
# now that we change some of the data types, let us check again with the info() method
data_frame.info()
```

#### Exercise 4.03 – working with text data using string methods

```python
import pandas as pd
```

```python
# let us create a series with some text data
s = pd.Series(
    ['          Data Analysis using python with pandas is great',
                'pandas DataFrame and pandas series are useful   ',
                'PYTHON3 PANDAS']
)
s
```

```python
# using the count() fucntions to count the number of times the word "pandas" appears in the text stored in s, 
# you'll see that pandas is case sensitive
s.str.count('pandas')
```

```python
# counting the numbers of characters in each string
s.str.len()
```

```python
# in the prevous result we got higher numbers due to leading blank spaces
# let us fix this, and at the same time, we'll get the new lengths
# observe that we need to "invoke" again the "sts" attribute
s.str.strip().str.len()
```

```python
# let us check if the the strings start with 'p'
s.str.startswith('p')
```

```python
# to make every string to start with "p" we will use a combination of functions and methods
s.str.strip().str.replace('Data', 'pandas').str.lower().str.startswith('p')
```

```python
# Use the get_dummies() function to create a DataFrame of dummy variables
# Dummy variables are Booleans that take the True/1 value when the column name is contained in the row.
s.str.get_dummies(' ')
```

```python
# Use a combination of functions to create a DataFrame of dummy variables that contains only pandas and python:
s = s.str.strip().str.lower()
s.str.get_dummies(' ')[['python', 'pandas']]
```
